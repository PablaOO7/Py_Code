### 7. Create a function to extract all data belonging to any car model in the dataset and export all the relevant stats separately.

```python
def car(model_name):
    info = mt[mt["model"] == model_name]
    return info

car("Duster 360")
```

### 8. Make a scatter plot visualization and comment on the correlation of the two variables.

```python
import plotly.express as px
fig = px.scatter_matrix(mt,
    dimensions=['mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'],
                        symbol="model",title="Interactive Scatter Plot of mtCARS",color="model")
fig.update_layout(width=1000,height=1200,title='Interactive Scatter Plot of mtCARS')
fig.show()
```

### 9. Make a heatmap on the correlation matrix of this dataframe “mt”.

```python
import matplotlib.pyplot as plot
import plotly.graph_objects as go
import cufflinks as cf
import plotly.offline
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
mt.corr().iplot(kind='heatmap',colorscale="Spectral",title="Feature Correlation Matrix")
plot.savefig('heatmap.png')
```
